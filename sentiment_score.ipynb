```python
from googletrans import Translator
import pandas as pd
import numpy as np
```

```python
annual_pubs_df = pd.read_csv("annual_posts.csv")
annual_pubs_df.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
```

```python
def sentiment_analyzer_scores(sentence):
    score = analyzer.polarity_scores(sentence)
    return score
```

```python
translated = pd.read_csv("translations.csv")
translated.drop(columns = "Unnamed: 0", inplace = True)

# test = sentiment_analyzer_scores(translated["0"][15])
# translated["score"] = annual_pubs_df.Message.apply(lambda x: sentiment_analyzer_scores(x))
# translated
translated["score"] = ""
for i, new in enumerate(translated["0"]):
    dict_score = sentiment_analyzer_scores(new)
    translated["score"][i] = dict_score["compound"]
    
translated.rename(columns = {"0" : "message"}, inplace = True)

def sentiment(score):
    if score >= 0.05:
        # Positive sentiment
        return 3
    elif score > -0.05 and score < 0.05:
        # Neutral 
        return 2
    else:
        #Negative
        return 1

translated["sentiment"] = translated.score.apply(lambda x: sentiment(x))
translated
translated.to_csv("translated_scores.csv", index = False)
```
